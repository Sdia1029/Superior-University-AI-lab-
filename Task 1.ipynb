```python
class ToDoList:
    def __init__(self):
        self.tasks = []  

    def add_task(self):
        task_name = input("Enter the task you want to add: ")
        while True:
            try:
                priority = int(input("Enter the priority (1-5, 1 being the highest): "))
                if 1 <= priority <= 5:
                    break
                else:
                    print("Priority must be between 1 and 5.")
            except ValueError:
                print("Invalid input. Please enter a number between 1 and 5.")
        self.tasks.append((task_name, priority))
        self.sort_tasks_by_priority()
        print(f"Task '{task_name}' added successfully with priority {priority}.")

    def update_task(self):
        old_task = input("Enter the task you want to update: ")
        index = -1
        
        for i in range(len(self.tasks)):
            if self.tasks[i][0] == old_task:
                index = i
                break
        
        if index != -1:
            new_task = input("Enter the new task : ")
            if new_task == "":
                new_task = self.tasks[index][0]
            while True:
                try:
                    new_priority = int(input("Enter the new priority (1-5): "))
                    if 1 <= new_priority <= 5:
                        break
                    else:
                        print("Priority must be between 1 and 5.")
                except ValueError:
                    new_priority = self.tasks[index][1]
                    print("Invalid input. Keeping old priority.")
            
            self.tasks[index] = (new_task, new_priority)
            self.sort_tasks_by_priority()
            print(f"Task '{old_task}' updated to '{new_task}' with priority {new_priority}.")
        else:
            print("Task not found.")

    def delete_task(self):
        del_task = input("Enter the task you want to delete: ")
        for i in range(len(self.tasks)):
            if self.tasks[i][0] == del_task:
                del self.tasks[i]
                print(f"Task '{del_task}' has been deleted.")
                return
        print("Task not found.")

    def view_tasks(self):
        if self.tasks:
            print("Today's Tasks :")
            for task, priority in self.tasks:
                print(f"Task: {task} , Priority: {priority}")
        else:
            print("No tasks to show.")

    def get_priority(self, task):
        return task[1]  

    def sort_tasks_by_priority(self):
        self.tasks.sort(key=self.get_priority)  

    def manage_tasks(self):
        while True:
            print("""
              1. Add Task
              2. Update Task
              3. Delete Task
              4. View Tasks
              5. Exit
            """)
            choice = input("Enter your choice: ")
            if choice == "1":
                self.add_task()
            elif choice == "2":
                self.update_task()
            elif choice == "3":
                self.delete_task()
            elif choice == "4":
                self.view_tasks()
            elif choice == "5":
                print("Exiting...")
                break
            else:
                print("Invalid choice. Please try again.")

manager = ToDoList()
manager.manage_tasks()
```
